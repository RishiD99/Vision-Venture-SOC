```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])   #this is for making legend
plt.show()
```

```python
x1 = np.arange(0, 2*np.pi, 0.1)
y_sin1 = np.sin(x1)
y_cos1 = np.cos(x1)

plt.subplot(2, 1, 1) #First subplot
plt.plot(x1, y_sin1)

plt.subplot(2, 1, 2) #Second subplot
plt.plot(x1, y_cos1)

plt.show()
```
